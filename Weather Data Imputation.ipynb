# Read files related to train data

```python
weather_train     = pd.read_csv('weather_train.csv')
```

# Functions

```python
#missing value computation
def cal_missing_val(df):
    data_dict = {}
    for col in df.columns:
        data_dict[col] = df[col].isnull().sum()/df.shape[0]*100
    return pd.DataFrame.from_dict(data_dict, orient='index', columns=['MissingValue'])
```

## EDA - Missing value 

```python
missing_weather  = cal_missing_val(weather_train)
```

## Imputing missing values by replacing them with the mean of the column

```python
#weather_train["air_temperature_mean"]=weather_train["air_temperature"].fillna(weather_train["air_tempreature"].mean())
#No idea why the air_temprature is not working
weather_train['cloud_coverage_mean']=weather_train['cloud_coverage'].fillna(weather_train['cloud_coverage'].mean())
weather_train['dew_temperature_mean']=weather_train['dew_temperature'].fillna(weather_train['dew_temperature'].mean())
weather_train['precip_depth_1_hr_mean']=weather_train['precip_depth_1_hr'].fillna(weather_train['precip_depth_1_hr'].mean())
weather_train['sea_level_pressure_mean']=weather_train['sea_level_pressure'].fillna(weather_train['sea_level_pressure'].mean())
weather_train['wind_direction_mean']=weather_train['wind_direction'].fillna(weather_train['wind_direction'].mean())
weather_train['wind_speed_mean']=weather_train['wind_speed'].fillna(weather_train['wind_speed'].mean())
weather_train.head(2)
```

## Imputing missing values by replacing them with the mean of the column per site_id

```python
weather_train['air_temperature_gb_mean']=weather_train['air_temperature'].fillna(weather_train.groupby(['site_id'])['air_temperature'].transform(np.mean))
weather_train['cloud_coverage_gb_mean']=weather_train['cloud_coverage'].fillna(weather_train.groupby(['site_id'])['cloud_coverage'].transform(np.mean))
weather_train['dew_temperature_gb_mean']=weather_train['dew_temperature'].fillna(weather_train.groupby(['site_id'])['dew_temperature'].transform(np.mean))
weather_train['precip_depth_1_hr_gb_mean']=weather_train['precip_depth_1_hr'].fillna(weather_train.groupby(['site_id'])['precip_depth_1_hr'].transform(np.mean))
weather_train['sea_level_pressure_gb_mean']=weather_train['sea_level_pressure'].fillna(weather_train.groupby(['site_id'])['sea_level_pressure'].transform(np.mean))
weather_train['wind_direction_gb_mean']=weather_train['wind_direction'].fillna(weather_train.groupby(['site_id'])['wind_direction'].transform(np.mean))
weather_train['wind_speed_gb_mean']=weather_train['wind_speed'].fillna(weather_train.groupby(['site_id'])['wind_speed'].transform(np.mean))
weather_train.head(2)
```

## Imputing missing values by replacing them with the rolling mean(7days)

```python
weather_train['air_temperature_R7'] = weather_train['air_temperature'].rolling( 7,center=True,min_periods=1).mean()
weather_train['air_temperature_R7'].update( weather_train['air_temperature'] )

weather_train['cloud_coverage_R7'] = weather_train['cloud_coverage'].rolling( 7,center=True,min_periods=1).mean()
weather_train['cloud_coverage_R7'].update( weather_train['cloud_coverage'] )

weather_train['dew_temperature_R7'] = weather_train['dew_temperature'].rolling( 7,center=True,min_periods=1).mean()
weather_train['dew_temperature_R7'].update( weather_train['dew_temperature'] )

weather_train['precip_depth_1_hr_R7'] = weather_train['precip_depth_1_hr'].rolling( 7,center=True,min_periods=1).mean()
weather_train['precip_depth_1_hr_R7'].update( weather_train['precip_depth_1_hr'] )

weather_train['sea_level_pressure_R7'] = weather_train['sea_level_pressure'].rolling( 7,center=True,min_periods=1).mean()
weather_train['sea_level_pressure_R7'].update( weather_train['sea_level_pressure'] )

weather_train['wind_direction_R7'] = weather_train['wind_direction'].rolling( 7,center=True,min_periods=1).mean()
weather_train['wind_direction_R7'].update( weather_train['wind_direction'] )

weather_train['wind_speed_R7'] = weather_train['wind_speed'].rolling( 7,center=True,min_periods=1).mean()
weather_train['wind_speed_R7'].update( weather_train['wind_speed'] )

weather_train.head(2)
```
