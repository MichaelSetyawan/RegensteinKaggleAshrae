```python
#Check the missing values in each table and dataframe it
def cal_missing_values(df):
    data_dict = {}
    for i in df.columns:
        data_dict[i] = df[i].isnull().sum()
    Dataframe = pd.DataFrame.from_dict(data_dict, orient = "index", columns = ['MissingValues'])
    return (Dataframe)
```

```python
# Adding columns of mean, median, max and min of air temperature to replace the missing values
weather_train['air_temperature_mean'] = weather_train['air_temperature']
weather_train['air_temperature_median'] = weather_train['air_temperature']
weather_train['air_temperature_max'] = weather_train['air_temperature']
weather_train['air_temperature_min'] = weather_train['air_temperature']
weather_train['air_temperature_mean'] = weather_train['air_temperature_mean'].fillna(np.mean(weather_train.air_temperature))
weather_train['air_temperature_median'] = weather_train['air_temperature_mean'].fillna(np.median(weather_train.air_temperature))
weather_train['air_temperature_max'] = weather_train['air_temperature_mean'].fillna(np.max(weather_train.air_temperature))
weather_train['air_temperature_min'] = weather_train['air_temperature_mean'].fillna(np.min(weather_train.air_temperature))
print(weather_train)
```

```python
# Adding columns of mean, median, max and min of dew temperature to replace the missing values
weather_train['dew_temperature_mean'] = weather_train['dew_temperature']
weather_train['dew_temperature_median'] = weather_train['dew_temperature']
weather_train['dew_temperature_max'] = weather_train['dew_temperature']
weather_train['dew_temperature_min'] = weather_train['dew_temperature']
weather_train['dew_temperature_mean'] = weather_train['dew_temperature_mean'].fillna(np.mean(weather_train.dew_temperature))
weather_train['dew_temperature_median'] = weather_train['dew_temperature_median'].fillna(np.median(weather_train.dew_temperature))
weather_train['dew_temperature_max'] = weather_train['dew_temperature_max'].fillna(np.max(weather_train.dew_temperature))
weather_train['dew_temperature_min'] = weather_train['dew_temperature_min'].fillna(np.min(weather_train.dew_temperature))
print(weather_train)
```

```python
# Adding columns of mean, median, max and min of cloud coverage to replace the missing values
weather_train['cloud_coverage_mean'] = weather_train['cloud_coverage']
weather_train['cloud_coverage_median'] = weather_train['cloud_coverage']
weather_train['cloud_coverage_max'] = weather_train['cloud_coverage']
weather_train['cloud_coverage_min'] = weather_train['cloud_coverage']
weather_train['cloud_coverage_mean'] = weather_train['cloud_coverage_mean'].fillna(np.mean(weather_train.cloud_coverage))
weather_train['cloud_coverage_median'] = weather_train['cloud_coverage_median'].fillna(np.median(weather_train.cloud_coverage))
weather_train['cloud_coverage_max'] = weather_train['cloud_coverage_max'].fillna(np.max(weather_train.cloud_coverage))
weather_train['cloud_coverage_min'] = weather_train['cloud_coverage_min'].fillna(np.min(weather_train.cloud_coverage))
print(weather_train)
```
